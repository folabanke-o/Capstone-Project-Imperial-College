```python
#A: Load the Initial Data

import numpy as np

# Dictionary to store all datasets
data = {}

# Load all 8 functions
for i in range(1, 9):

    X = np.load(f"function_{i}/initial_inputs.npy")
    Y = np.load(f"function_{i}/initial_outputs.npy")

    data[f"function_{i}"] = {
        "X": X,
        "Y": Y
    }

    print(f"\nFunction {i}")
    print("Input shape:", X.shape)
    print("Output shape:", Y.shape)
```

```python
#LOAD AND INSPECT FUNCTION 1 DATA
#Access Function 1 input and output arrays
X1 = data["function_1"]["X"]
Y1 = data["function_1"]["Y"]

# Display dataset dimensions
print("Function 1 Dataset Overview")
print("-" * 40)

print(f"Input matrix shape : {X1.shape}")
print(f"Output vector shape: {Y1.shape}")

# Display first few samples
print("\nFirst 5 input samples:")
print(X1[:5])

print("\nFirst 5 output values:")
print(Y1[:5])
```

```python
#B2: OUTPUT STATISTICAL ANALYSIS FUNCTION 1 DATA
#Calculate key statistical properties of the outputs

minimum_output = Y1.min()
maximum_output = Y1.max()
mean_output = Y1.mean()
std_output = Y1.std()

# Display statistical summary
print("Function 1 Output Statistics")
print("-" * 40)

print(f"Minimum output       : {minimum_output:.6f}")
print(f"Maximum output       : {maximum_output:.6f}")
print(f"Mean output          : {mean_output:.6f}")
print(f"Standard deviation   : {std_output:.6f}")
```

```python
# FIND THE CURRENT BEST OBSERVATION FUNCTION 1
# Locate index of highest observed output
best_index = np.argmax(Y1)

# Retrieve corresponding input and output
best_input = X1[best_index]
best_output = Y1[best_index]

# Display best observed point
print("Best Current Observation")
print("-" * 40)

print(f"Best input point : {best_input}")
print(f"Best output value: {best_output:.6f}")
```

```python
#VISUALISE FUNCTION 1 SAMPLE DISTRIBUTION FUNCTION 1 DATA

import matplotlib.pyplot as plt

# Create figure
plt.figure(figsize=(8, 6))

# Scatter plot of sampled points
scatter = plt.scatter(
    X1[:, 0],              # x-axis values
    X1[:, 1],              # y-axis values
    c=Y1,                  # colour based on output value
    s=120,                 # marker size
    edgecolor='black'
)

# Add colour bar
cbar = plt.colorbar(scatter)
cbar.set_label("Function Output", fontsize=12)

# Label axes
plt.xlabel("x1", fontsize=12)
plt.ylabel("x2", fontsize=12)

# Add title
plt.title("Function 1 — Initial Sample Distribution", fontsize=14)

# Add grid
plt.grid(True, alpha=0.3)

# Show plot
plt.show()
```
